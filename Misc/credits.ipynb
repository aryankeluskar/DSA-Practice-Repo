```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

credits = {"AP": 26, "1": 17, "2": 19, "3": 23, "4": 20, "5": 18, "6": 15}
df = pd.DataFrame(list(credits.items()), columns=['Semester', 'Credits'])
# print(df)

plt.plot(df['Semester'], df['Credits'], marker='o')
plt.xlabel('Semester')
plt.ylabel('Credits')
plt.title('Credits Over Time')

plt.show()
```

```python
# add a line for cumulative credits
cumulative = np.cumsum(df['Credits'])
plt.plot(df['Semester'], cumulative, marker='o')

# add y value beside each point, offset by y = 5 to avoid overlap
for i, txt in enumerate(cumulative):
    if i < 6:
        plt.annotate(txt, (df['Semester'][i], cumulative[i] + 5))
plt.annotate(cumulative[6], (df['Semester'][6], cumulative[6] - 8))

# add histogram for credits per semester, with value on top of each bar
plt.bar(df['Semester'], df['Credits'], color='#8C1D40')
for i, v in enumerate(df['Credits']):
    if i > 0:
        plt.text(i-0.1, v + 1.5, str(v), color='black')

# add a horizontal line for y = 120, label as "Graduation Requirement"
plt.axhline(y=120, color='r', linestyle='--')
plt.text(0, 122, 'Graduation Requirement = 120', color='r')

plt.axhline(y=25, color='g', linestyle='--', alpha=0.5)
plt.text(4.5, 27, 'Sophomore = 25', color='g')
plt.axhline(y=55, color='g', linestyle='--')
plt.text(4.5, 58, 'Junior = 56', color='g')
plt.axhline(y=85, color='g', linestyle='--')
plt.text(4.5, 89, 'Senior = 87', color='g')

# add legend saying #8C1D40 box is credits per semester, and line is cumulative credits
plt.legend(['Cumulative Credits'], loc='upper left')

plt.xlabel('Semester')
plt.ylabel('Cumulative Credits')
plt.title('Cumulative Credits Over Time')

plt.savefig('credits.png')
```
